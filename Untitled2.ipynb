# Attribute Information

##### Name : Name of the video game
##### Platform : In which platform the game will be available
##### Year_of_Release : In which year the game was released
##### Genre : Shows which type of game is it
##### Publisher : Tells who is the publisher
##### NA_Sales : Shows the NASALES of the data
##### EU_Sales : Shows the EUSALES of the data
##### JP_Sales : Shows the JPSALES of the data
##### Other_Sales : Shows the other sales
##### Global_Sales : Shows the Global Sales of the data
##### Critic_Score : Shows the critic score of the data
##### Critic_Count : Shows the critic count of the data
##### User_Score : Shows the user score 
##### User_Count : Tells the user count
##### Developer : Shows the developer of the game
##### Rating : Shows the rating of the game
